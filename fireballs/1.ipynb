```python
import pandas as pd
df = pd.read_csv('nasas fireballs.csv')
df.head(10)

def cleanCoordinate(coordinate):
    c = coordinate[-1]
    if c =='N': coordinate = f"+{coordinate[:-1]}"
    if c =='S': coordinate = f"-{coordinate[:-1]}"
    if c =='E': coordinate = f"+{coordinate[:-1]}"
    if c =='W': coordinate = f"-{coordinate[:-1]}"
    return float(coordinate)
cleanCoordinate('55.5S')

df = df.drop_duplicates()
df = df.dropna()


df['latitude'] = df['latitude'].apply(cleanCoordinate)
df['longitude'] = df['longitude'].apply(cleanCoordinate)

df['velocity(km/s)'] = abs(df['velocity(km/s)'])


# sum(df['total impact energy'].isnull() == True)

df2 = pd.DataFrame()
dates = df['date/time for peak brightness'].sort_values(ascending = True).tolist()

for date in dates:
    # tdf = df[df['date/time for peak brightness'] <= date]
    tdf = df.loc[df['date/time for peak brightness'] <= date]
    tdf['date/time for peak brightness'] = date
    df2 = pd.concat([df2,tdf], axis = 0)



df2 = df2.sort_values(by=['date/time for peak brightness'],ascending = True)
df2 = df2.reset_index()
df2 = df2.drop(columns=['index'], axis=1)
df2
```

```python
import plotly.express as px
fig = px.scatter_mapbox(df2, lat="latitude", lon="longitude", 
color="total impact energy",
range_color=[min(df2['total impact energy']),max(df2['total impact energy'])],
size="velocity(km/s)",
size_max=10,
labels='Asteroids',
animation_frame='date/time for peak brightness'
)
fig.update_layout(mapbox_style="carto-positron")

fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 50


import plotly.io as pio
pio.renderers.default = 'notebook_connected'

fig.show(renderer='notebook_connected')
```
