```python
import pandas as pd
import yfinance as yf
import plotly.express as px
st = '2021-03-01'
end = '2022-04-20'


data = pd.read_csv('constituents.csv')
symbol = data["Symbol"].tolist()[:10]
sector = data["Sector"].tolist()[:10]

df = pd.DataFrame()
for i in range(len(symbol)):
    removed = []
    try:
        a = yf.download([f'{symbol[i]}'], 
                            start=st, 
                            end=end, 
                            progress=False,
                            interval = '1wk'
        )
        a['Ticker'] = symbol[i]
        a['Area'] = sector[i]
        a['Rate'] = (100*a['Close']/a['Close'][0])-100
        # df = df.append(a)
        df = pd.concat([df, a], axis=0)
        print("download number: #",i )
    except:
        removed = removed.append(symbol[i])


def cut(x):
    return str(x)[:-9]

df['date'] = df.index
df.date = df.date.apply(cut)
df.dropna()
df.head()
```

```python
import plotly.express as px

fig = px.scatter(df, x = df['Volume'], y = df.Rate, animation_frame=df.date, animation_group=df.Ticker,
           size=[1]*len(df['Volume']), color=df.Area, hover_name=df.Ticker,
           log_x=True, size_max=40, range_x=[100,1000000000], range_y=[000,1500])
           
fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 100


import plotly.io as pio
pio.renderers.default = 'notebook_connected'

fig.show(renderer='notebook_connected')
```
